```python
import os
import sys
import json
import cv2
import numpy as np
import torch
import random
import warnings
import urllib.parse
import matplotlib.pyplot as plt
import gc
import logging
import concurrent.futures
from pathlib import Path
from dataclasses import dataclass, field
from tqdm.auto import tqdm
from sklearn.model_selection import train_test_split
from functools import partial

## DEPENDENCY INSTALLATION
print("⚙️ Setting up Environment...")
os.system('pip uninstall -y opencv-python opencv-contrib-python opencv-python-headless')
os.system('pip install -q --no-deps opencv-python-headless==4.10.0.84') 
os.system('pip install -q --no-deps "numpy<2.0" "ultralytics>=8.3.0" segmentation-models-pytorch albumentations torchmetrics torchinfo')

import segmentation_models_pytorch as smp
from ultralytics import YOLO
import albumentations as albu
from albumentations.pytorch import ToTensorV2
import torchmetrics
from torch.utils.data import DataLoader, Dataset
from torch.cuda.amp import autocast, GradScaler
import torch.nn as nn
import torch.optim as optim

## CONFIGURATION
@dataclass
class Config:
    RAW_JSON: Path = Path("/kaggle/input/patch-placement/Work/project-4-at-2025-11-19-03-05-8b131c6a.json")
    RAW_IMG_DIR: Path = Path("/kaggle/input/patch-placement/Work/InnerGize/Datasets/Device_Placement")
    WORK_DIR: Path = Path("/kaggle/working/dataset_512_aug_v2")
    ARCH: str = 'UnetPlusPlus'
    ENCODER: str = 'efficientnet-b4'
    INPUT_SIZE: int = 512
    BATCH_SIZE: int = 8
    ACCUM_STEPS: int = 4
    LR: float = 3e-4
    EPOCHS: int = 150       
    PATIENCE: int = 25      
    DEVICE: str = 'cuda' if torch.cuda.is_available() else 'cpu'
    SEED: int = 42
    IMG_OUT: Path = field(init=False)
    MASK_OUT: Path = field(init=False)
    SPLIT_DIR: Path = field(init=False)

    def __post_init__(self):
        self.IMG_OUT = self.WORK_DIR / "images"
        self.MASK_OUT = self.WORK_DIR / "masks"
        self.SPLIT_DIR = self.WORK_DIR / "splits"
        os.makedirs(self.IMG_OUT, exist_ok=True)
        os.makedirs(self.MASK_OUT, exist_ok=True)
        os.makedirs(self.SPLIT_DIR, exist_ok=True)

CFG = Config()

## LOGGING SETUP
def setup_logger(cfg: Config):
    logger = logging.getLogger("KagglePipeline")
    logger.setLevel(logging.INFO)
    if logger.hasHandlers(): logger.handlers.clear()
    formatter = logging.Formatter('%(asctime)s | %(levelname)s | %(message)s', datefmt='%H:%M:%S')
    sh = logging.StreamHandler(sys.stdout); sh.setFormatter(formatter)
    fh = logging.FileHandler(cfg.WORK_DIR / "pipeline.log"); fh.setFormatter(formatter)
    logger.addHandler(sh); logger.addHandler(fh)
    return logger

LOGGER = setup_logger(CFG)

def seed_everything(seed=42):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)

seed_everything(CFG.SEED)
warnings.filterwarnings("ignore")

## MULTIPROCESSING WORKER
def process_single_item(item, cfg, person_masks):
    try:
        url = item['image']
        if "?d=" in url: url = url.split("?d=")[1]
        clean_url = urllib.parse.unquote(url).replace("\\", "/")
        fname = Path(clean_url).name
        img_path = cfg.RAW_IMG_DIR / fname
        item_id = str(item.get('id', fname))

        if not img_path.exists() or item_id not in person_masks:
            return False

        img_bgr = cv2.imread(str(img_path))
        if img_bgr is None: return False
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
        h, w = img_rgb.shape[:2]

        m = person_masks[item_id]
        person_mask = (cv2.resize(m, (w, h)) > 0.5).astype(np.uint8)

        gt_mask_full = np.zeros((h, w), dtype=np.uint8)
        device_points = []
        if 'label' in item and item['label']:
            lbl = item['label'][0]
            pts = np.array(lbl['points'])
            pts[:, 0] *= (lbl['original_width'] / 100.0)
            pts[:, 1] *= (lbl['original_height'] / 100.0)
            device_points = pts
            cv2.fillPoly(gt_mask_full, [pts.astype(np.int32)], 1)

        black_bg = np.zeros_like(img_rgb)
        clean_img = np.where(np.stack((person_mask,) * 3, axis=-1) > 0, img_rgb, black_bg)

        y_head_top = np.argmax(np.any(person_mask, axis=1)) if np.any(person_mask) else 0
        y_device_bottom = int(np.max(device_points[:, 1])) if len(device_points) > 0 else h
        roi_height = max(y_device_bottom - y_head_top, h // 3)
        square_dim = max(int(roi_height * 1.5), 256)

        mask_slice = person_mask[max(0, y_head_top):min(h, y_device_bottom), :]
        if np.sum(mask_slice) > 0:
            center_x = int(np.dot(np.arange(w), np.sum(mask_slice, axis=0)) / np.sum(mask_slice))
        else:
            center_x = w // 2
        center_y = y_head_top + (roi_height // 2)

        half = square_dim // 2
        x1, y1 = center_x - half, center_y - half
        pad_l, pad_t = max(0, -x1), max(0, -y1)
        pad_r, pad_b = max(0, (x1 + square_dim) - w), max(0, (y1 + square_dim) - h)

        padded_img = cv2.copyMakeBorder(clean_img, pad_t, pad_b, pad_l, pad_r, cv2.BORDER_CONSTANT, value=[0,0,0])
        padded_mask = cv2.copyMakeBorder(gt_mask_full, pad_t, pad_b, pad_l, pad_r, cv2.BORDER_CONSTANT, value=0)

        cx1, cy1 = x1 + pad_l, y1 + pad_t
        fin_img = cv2.resize(padded_img[cy1:cy1+square_dim, cx1:cx1+square_dim], (cfg.INPUT_SIZE, cfg.INPUT_SIZE))
        fin_mask = cv2.resize(padded_mask[cy1:cy1+square_dim, cx1:cx1+square_dim], (cfg.INPUT_SIZE, cfg.INPUT_SIZE), interpolation=cv2.INTER_NEAREST)

        out_name = Path(fname).stem + f"_{item_id}.png"
        cv2.imwrite(str(cfg.IMG_OUT / out_name), cv2.cvtColor(fin_img, cv2.COLOR_RGB2BGR))
        cv2.imwrite(str(cfg.MASK_OUT / out_name), fin_mask * 255)
        return True
    except Exception:
        return False

## PREPROCESSING ENGINE
class Preprocessor:
    def __init__(self, cfg: Config):
        self.cfg = cfg

    def process_data(self):
        if len(list(self.cfg.IMG_OUT.iterdir())) < 5:
            LOGGER.info("⬇️ Starting Preprocessing...")
            self._generate_images()
        self._generate_splits()

    def _generate_images(self):
        yolo = YOLO('yolo11n-seg.pt')
        with open(self.cfg.RAW_JSON, 'r') as f: data = json.load(f)
        valid_items = [x for x in data if x.get('image')]
        
        person_masks = {}
        for item in tqdm(valid_items, desc="YOLO Masking"):
            try:
                url = item['image']
                if "?d=" in url: url = url.split("?d=")[1]
                fname = Path(urllib.parse.unquote(url)).name
                img_path = self.cfg.RAW_IMG_DIR / fname
                item_id = str(item.get('id', fname))
                if not img_path.exists(): continue
                results = yolo(str(img_path), verbose=False, retina_masks=False)
                if results[0].masks is not None:
                    cls = results[0].boxes.cls
                    if 0 in cls:
                        idx = (cls == 0).nonzero(as_tuple=True)[0][0] 
                        person_masks[item_id] = results[0].masks.data[idx].cpu().numpy()
            except Exception: continue
            
        del yolo; gc.collect(); torch.cuda.empty_cache()
        LOGGER.info(f"Multi-threaded Cropping...")
        with concurrent.futures.ProcessPoolExecutor(max_workers=4) as executor:
            worker_fn = partial(process_single_item, cfg=self.cfg, person_masks=person_masks)
            results = list(tqdm(executor.map(worker_fn, valid_items), total=len(valid_items), desc="Cropping"))
        LOGGER.info(f"Preprocessed: {sum(results)}")

    def _generate_splits(self):
        files = sorted([f.name for f in self.cfg.IMG_OUT.iterdir() if f.suffix == '.png'])
        train, val = train_test_split(files, test_size=0.15, random_state=42)
        np.save(self.cfg.SPLIT_DIR / "train.npy", train)
        np.save(self.cfg.SPLIT_DIR / "val.npy", val)

## DATASET
class SegmentationDataset(Dataset):
    def __init__(self, file_list, cfg, transform=None):
        self.files = file_list
        self.cfg = cfg
        self.transform = transform
    def __len__(self): return len(self.files)
    def __getitem__(self, idx):
        fname = self.files[idx]
        image = cv2.cvtColor(cv2.imread(str(self.cfg.IMG_OUT / fname)), cv2.COLOR_BGR2RGB)
        mask = cv2.imread(str(self.cfg.MASK_OUT / fname), cv2.IMREAD_GRAYSCALE)
        if self.transform:
            res = self.transform(image=image, mask=mask)
            image, mask = res['image'], res['mask']
        return image, mask.unsqueeze(0).float() / 255.0

def get_transforms(phase):
    base = [albu.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)), ToTensorV2()]
    if phase == 'train':
        return albu.Compose([
            albu.HorizontalFlip(p=0.5),
            albu.ShiftScaleRotate(scale_limit=0.15, rotate_limit=20, p=0.7),
            albu.RandomBrightnessContrast(p=0.4),
        ] + base)
    return albu.Compose(base)

## TRAINER
class Trainer:
    def __init__(self, cfg):
        self.cfg = cfg
        train_f = np.load(cfg.SPLIT_DIR / "train.npy")
        val_f = np.load(cfg.SPLIT_DIR / "val.npy")
        self.train_loader = DataLoader(SegmentationDataset(train_f, cfg, get_transforms('train')), batch_size=cfg.BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
        self.val_loader = DataLoader(SegmentationDataset(val_f, cfg, get_transforms('val')), batch_size=cfg.BATCH_SIZE, num_workers=2, pin_memory=True)
        
        self.model = smp.UnetPlusPlus(encoder_name=cfg.ENCODER, encoder_weights="imagenet", in_channels=3, classes=1).to(cfg.DEVICE)
        self.optimizer = optim.AdamW(self.model.parameters(), lr=cfg.LR)
        self.criterion = smp.losses.DiceLoss(mode='binary', from_logits=True)
        self.metric = torchmetrics.JaccardIndex(task="binary").to(cfg.DEVICE)
        self.scaler = GradScaler()
        self.history = {'train_loss': [], 'val_loss': [], 'val_iou': []}

    def run(self):
        best_iou = 0.0
        patience = 0
        overall_pbar = tqdm(range(self.cfg.EPOCHS), desc="Overall Training")
        
        for epoch in overall_pbar:
            self.model.train()
            train_loss = 0
            epoch_pbar = tqdm(self.train_loader, desc=f"Epoch {epoch+1}", leave=False)
            for imgs, masks in epoch_pbar:
                imgs, masks = imgs.to(self.cfg.DEVICE, non_blocking=True), masks.to(self.cfg.DEVICE, non_blocking=True)
                with autocast():
                    preds = self.model(imgs)
                    loss = self.criterion(preds, masks)
                self.scaler.scale(loss).backward()
                self.scaler.step(self.optimizer); self.scaler.update(); self.optimizer.zero_grad()
                train_loss += loss.item()
                epoch_pbar.set_postfix(t_loss=loss.item())
            
            self.model.eval(); self.metric.reset()
            val_loss = 0
            with torch.no_grad():
                for imgs, masks in self.val_loader:
                    imgs, masks = imgs.to(self.cfg.DEVICE, non_blocking=True), masks.to(self.cfg.DEVICE, non_blocking=True)
                    preds = self.model(imgs)
                    v_l = self.criterion(preds, masks)
                    val_loss += v_l.item()
                    self.metric.update(preds.sigmoid() > 0.5, masks.int())
            
            avg_t_loss = train_loss / len(self.train_loader)
            avg_v_loss = val_loss / len(self.val_loader)
            iou = self.metric.compute().item()
            
            self.history['train_loss'].append(avg_t_loss)
            self.history['val_loss'].append(avg_v_loss)
            self.history['val_iou'].append(iou)
            
            overall_pbar.set_postfix(iou=f"{iou:.4f}", t_loss=f"{avg_t_loss:.3f}", v_loss=f"{avg_v_loss:.3f}")
            LOGGER.info(f"Ep {epoch+1} | IoU: {iou:.4f} | Train Loss: {avg_t_loss:.4f} | Val Loss: {avg_v_loss:.4f}")

            if iou > best_iou:
                best_iou = iou; patience = 0
                torch.save(self.model.state_dict(), self.cfg.WORK_DIR / "best_model.pth")
            else:
                patience += 1
                if patience >= self.cfg.PATIENCE:
                    LOGGER.info("Early Stopping triggered")
                    break
        return self.history

## VISUALIZATION SUITE
def plot_results(history, trainer, cfg):
    # 1. Training Curves
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss', color='#e74c3c')
    plt.plot(history['val_loss'], label='Val Loss', color='#3498db')
    plt.title('Loss Curves (DiceLoss)')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history['val_iou'], label='Val IoU', color='#2ecc71', linewidth=2)
    plt.title('Metric Evolution (Jaccard Index)')
    plt.xlabel('Epoch'); plt.ylabel('IoU'); plt.legend()
    plt.tight_layout(); plt.show()

    # 2. Prediction Overlays
    trainer.model.load_state_dict(torch.load(cfg.WORK_DIR / "best_model.pth"))
    trainer.model.eval()
    imgs, masks = next(iter(trainer.val_loader))
    with torch.no_grad():
        preds = trainer.model(imgs.to(cfg.DEVICE)).sigmoid().cpu()
    
    count = min(3, len(imgs))
    fig, axes = plt.subplots(count, 3, figsize=(12, 4*count))
    mean = np.array([0.485, 0.456, 0.406]).reshape(3,1,1)
    std = np.array([0.229, 0.224, 0.225]).reshape(3,1,1)

    for i in range(count):
        img_np = (imgs[i].numpy() * std + mean).transpose(1, 2, 0).clip(0, 1)
        gt_np = masks[i].squeeze().numpy()
        pr_np = (preds[i].squeeze().numpy() > 0.5).astype(np.uint8)

        overlay = img_np.copy()
        overlay[gt_np > 0.5] = overlay[gt_np > 0.5] * 0.5 + np.array([0, 1, 0]) * 0.5 # GT in Green
        overlay[pr_np > 0.5] = overlay[pr_np > 0.5] * 0.5 + np.array([1, 0, 0]) * 0.5 # Pred in Red

        axes[i, 0].imshow(img_np); axes[i, 0].set_title("Original (Crop)")
        axes[i, 1].imshow(gt_np, cmap='gray'); axes[i, 1].set_title("Ground Truth")
        axes[i, 2].imshow(overlay); axes[i, 2].set_title("Red=Pred, Green=GT")
        for ax in axes[i]: ax.axis('off')
    
    plt.tight_layout(); plt.show()

## EXECUTION
if __name__ == "__main__":
    try:
        prep = Preprocessor(CFG)
        prep.process_data()
        trainer = Trainer(CFG)
        history = trainer.run()
        plot_results(history, trainer, CFG)
        LOGGER.info("Finished.")
    except Exception as e:
        LOGGER.error(f"Failure: {e}", exc_info=True)
```
